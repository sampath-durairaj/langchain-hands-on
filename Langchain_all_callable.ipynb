```python
def add_one(x):
    return x+1

#callable lambda function
add_two = lambda x: x + 2

print(add_one.__call__(4))
print(add_two.__call__(4))
```

```python
#Runnable (wrapping callable)
#convert Callable to Runnable
from langchain_core.runnables import RunnableLambda
runnable1 = RunnableLambda(add_one)

print(runnable1.invoke(5))
```

```python
#Runnable Passthrough - No action performed
#It will return the output as is input
# ex. If input is "Hello, World!" then output will be "Hello, World!"
#jsut pass data troguh without any modification
from langchain_core.runnables import RunnablePassthrough
chain = RunnablePassthrough()
my_input_data = {"input": "Hello World"}
result = chain.invoke(my_input_data)
print(result)  # Output: Hello, World!
```

```python
# this will make a dictionary with two keys 'lenth' and 'upper'
chain = RunnablePassthrough.assign(
    lenth=lambda x: len(x["input"]),
    upper = lambda x: x["input"].upper()
)

result = chain.invoke({"input":"python"})
print(result)  # Output: Hello, World!
```

```python
#Runnable Lambda

def make_uppercase(x: str) -> str:
    return x.upper()

uppercase_stpe = RunnableLambda(make_uppercase)
result = uppercase_stpe.invoke("hello world")
print(result)  # Output: HELLO WORLD
```

```python
#Runnable Lambda with chain
def add_excitment(x: str) -> str:
    return x + "!!!"

def concat(x: str, y: str) -> str:
    return x + y

chain =(
    RunnableLambda(make_uppercase) |
    RunnableLambda(add_excitment) |
    RunnableLambda(lambda x: concat(x, " Have a great day!") )
)
result = chain.invoke("hello world")
print(result)  # Output: HELLO WORLD
```

```python
#RunnableParallel - runs multiple runnables at the same time
#convert callables to runnables and run them in parallel
from langchain_core.runnables import RunnableParallel

def count_characters(x: str) -> int:
    return len(x)
def count_words(x: str) -> int:
    return len(x.split())
chain = RunnableParallel(
    words = RunnableLambda(count_words),
    characters = RunnableLambda(count_characters)   
)

input_text = "I Love Python programming"
result = chain.invoke(input_text)
print(result)  # Output: {'words': 4, 'characters': 24}
```
